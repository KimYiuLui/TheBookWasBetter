```python
import pandas as pd
```

```python
dirty_data = pd.read_csv("dataset\\books.csv")
```

```python
dirty_data.head()
```

```python
# Determining which columns to use and which to drop (because they are not used)
headers_to_keep = ['id', 'book_id', 'best_book_id', 'books_count', 'isbn', 'isbn13', 'authors', 'original_publication_year', 'original_title', 'average_rating', 'ratings_count', 'work_ratings_count', 'work_text_reviews_count', 'image_url']
dirty_data_good_headers = dirty_data[dirty_data.columns.intersection(headers_to_keep)]

dirty_data_good_headers.head()
```

```python
# Checking for Na/NaN and removing any, if present
dirty_data_good_headers.isnull().sum().sum()

# It may not show, but we have a few thousand-ish rows containing NaN. Considering we have a dataset of 10K, let's see where those NaN's are
for i in headers_to_keep:
    print(i)
    print(dirty_data_good_headers[i].isnull().values.any())
    print(dirty_data_good_headers[i].isnull().sum().sum())
    print("**********")

# Hmm. I'm taking a guess and say that the ISBN13 and original title NaN's are for the same rows, as it is very uncanny that they both have the same number
# Let's remove all NaN's and see with how many we end up with
print("New data with NaN removed")
dirty_data_remove_na = dirty_data_good_headers.dropna()

for i in headers_to_keep:
    print(i)
    print(dirty_data_remove_na[i].isnull().values.any())
    print(dirty_data_remove_na[i].isnull().sum().sum())
    print("**********")    

# We end up with over 8K rows left. Considering that is still 4/5 of the original dataset, I'll continue using the NaN-free dataset from now on
```

```python
# original_publication_year is not integer, so we must change that
publication_year_int = dirty_data_remove_na['original_publication_year']
publication_year_int = [int(i) for i in publication_year_int]

dirty_data_good_years = dirty_data_remove_na

# Now we insert the correct integer year into the df in place of the float years
dirty_data_good_years['original_publication_year'] = publication_year_int

# And now, to check
dirty_data_good_years.head()

# It works, yay ^.^
```

```python
# Nice, but those pesky isbn13 e+12 is bothering me. Let's see how we can change that
changing_isbn13 = dirty_data_good_years['isbn13']
changing_isbn13 = [int(i) for i in changing_isbn13]

# Ha, easy fix when you know the trick
dirty_data_good_years['isbn13'] = changing_isbn13

dirty_data_good_years.head()
```

```python
# Well, I have problems in the isbn row, so, for now, ive decided to remove that column, seeing as isbn13 is used more commonly anyway
dirty_data_no_isbn = dirty_data_good_years

dirty_data_no_isbn = dirty_data_no_isbn.drop(['isbn'], axis = 1)

dirty_data_no_isbn.head()
```

```python
# Nice, this looks right. I'm a bit concerned about those authors containing comma's, but let's see what happens. Firstly, let's give the df a final name change, for the prettiness of it all
clean_data = dirty_data_no_isbn
```
