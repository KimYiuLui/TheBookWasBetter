```python
import os

from PIL import Image
import requests
import urllib.request as urllib2
import shutil
from bs4 import BeautifulSoup
from io import BytesIO
import pandas as pd
import numpy as np
import os.path as path
import time
```

```python
# GLOBALS

OUTPUT_FOLDER = "output"
OUTPUT_CSV = "\output.csv"
```

```python
if(path.exists(OUTPUT_FOLDER + OUTPUT_CSV)):
    image_dataframe = pd.read_csv(OUTPUT_FOLDER + OUTPUT_CSV, error_bad_lines=False, decimal=",", header=0, dtype={'id':str})
    print("Previous output detected: Loading")
else:
    raise Exception("No input dataframe found. Stopping")
```

```python
def loop_through_df(image_dataframe):
    bookId = ""
    url = ""
    indexSinceError = 0
    
    for index in range(len(image_dataframe)):
        print(index)
        
        if(index < 50000): #Set limit, unless you want it to run 24/7, which I wouldn't.
            image_name = image_dataframe.iloc[index,0] #Id of book
            url = image_dataframe.iloc[index,2]

            if(path.exists(OUTPUT_FOLDER + "\\" + os.path.basename(image_name + ".jpg"))):
                print(image_name + " already present, skipping")
                continue

            response = requests.get(url)

            if(response.status_code != 200):
                if(index - indexSinceError > 100): #Amount of successfull attempts required until it errors again.
                    print("Blocked/Errored, waiting 5 minutes before retry")
                    time.sleep(300)
                    response = requests.get(url)
                    
                    if(response.status_code != 200):
                        print("Retry failed, stopping.")
                        break
        
                    indexSinceError = index;
                else:
                    print("Blocked too shortly after unblocked, stopping.")
                    break

            img = Image.open(BytesIO(response.content))
            img = img.convert('RGB');
            print(os.path.basename(image_name + ".jpg"))
            response = img.save(OUTPUT_FOLDER + "\\" + os.path.basename(image_name + ".jpg"))
            continue
            
        else: #Break after limit is reached.
            print("Limit reached, stopping")
            break
            
```

```python
loop_through_df(image_dataframe);
print("Done!")
```
